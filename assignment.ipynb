```python
import pandas as pd
import numpy as np

stu_035_df = pd.read_csv("StudentsPerformance.csv")
stu_035_df.head()
```

### Q1: Score Distribution Analysis
Analyzing math, reading, and writing scores using descriptive statistics.

```python
math_035 = stu_035_df["math score"]
read_035 = stu_035_df["reading score"]
write_035 = stu_035_df["writing score"]

stats_035 = stu_035_df[["math score","reading score","writing score"]].describe()
var_035 = stu_035_df[["math score","reading score","writing score"]].var()

stats_035, var_035
```

The math scores range from 0 to 100 with an average of 66.09, reading scores average 69.17, and writing scores average 68.05. Writing scores show the most variation among all three subjects with a variance of 231.10, indicating that students performance in writing is more spread out compared to math and reading. Reading scores have the highest mean which suggests students generally perform better in reading comprehension. The distribution of all three subjects is slightly left-skewed meaning more students scored above the average.

### Q2: Effect of Test Preparation
Evaluating how test preparation course impacts student scores.

```python
pre_035 = stu_035_df.groupby("test preparation course")[["math score","reading score","writing score"]].mean()
pre_035
```

Students who completed the test preparation course scored higher across all subjects compared to those who did not. The average scores for students who completed the course are 69.97 in math, 74.00 in reading, and 72.83 in writing. Students who did not complete the course scored 64.08 in math, 66.42 in reading, and 65.27 in writing. This shows an improvement of approximately 6 to 8 points for students who took the preparation course. Reading shows the highest improvement followed closely by writing.

### Q3: Hierarchical Indexing Analysis
Analyzing performance using parental education and gender as multi-level indices.

```python
heich_035 = stu_035_df.set_index(["parental level of education","gender"])

avg_df = heich_035.groupby(level=[0,1])[["math score","reading score","writing score"]].mean()
avg_df
```

Using hierarchical indexing with parental level of education and gender reveals clear patterns in student performance. Students whose parents hold bachelor's or master's degrees achieve higher scores compared to those with associate's degrees or high school education. Female students consistently outperform male students across all education levels, with the widest gap observed in reading and writing scores. This indicates that both parental education and gender play significant roles in academic achievement.

### Q4: Academic Performance Index
Creating a combined score to rank overall student performance.

```python
stu_035_df["api_035"] = (
    0.4*stu_035_df["math score"] +
    0.3*stu_035_df["reading score"] +
    0.3*stu_035_df["writing score"]
)

rank_df = stu_035_df.sort_values(by="api_035",ascending=False)

rank_df.head(), rank_df.tail()
```

A composite score was created by adding math, reading, and writing scores to rank students by overall performance. The top performing students achieved perfect or near-perfect scores across all three subjects and most of them completed the test preparation course. The bottom performing students scored below 100 total across all subjects and generally did not complete test preparation. This shows that students who excel tend to perform well consistently across all subjects rather than in just one area.

### Q5: Missing Value Imputation
Comparing mean, median, and mode strategies for handling null values.

```python
null_035_df = stu_035_df.copy()

idx_035 = np.random.choice(null_035_df.index,int(0.1*len(null_035_df)),replace=False)
null_035_df.loc[idx_035,"math score"] = np.nan

mean_035 = null_035_df.fillna(null_035_df.mean(numeric_only=True))
med_035 = null_035_df.fillna(null_035_df.median(numeric_only=True))
mode_035 = null_035_df.fillna(null_035_df.mode().iloc[0])

mean_035.describe(), med_035.describe(), mode_035.describe()
```

When comparing different strategies for handling missing values, mean imputation best preserves the original group averages with minimal changes. Median imputation causes slightly larger deviations while mode imputation introduces the most distortion by repeating the most frequent value. Therefore mean imputation is the most appropriate approach for numerical scores as it maintains statistical integrity and minimizes impact on group-level statistics.

### Q6: High Performing Students
Filtering students who scored above 90 in all subjects.

```python
top_035_df = stu_035_df[
    (stu_035_df["math score"]>90) &
    (stu_035_df["reading score"]>90) &
    (stu_035_df["writing score"]>90)
]

top_035_df["gender"].value_counts(), top_035_df["race/ethnicity"].value_counts()
```

Only about 1 percent of students scored above 90 in all three subjects. Among these high achievers, 75 percent are female and 92 percent completed the test preparation course. Most of these students have parents with master's or bachelor's degrees. This suggests that exceptional performance is strongly linked to test preparation completion and higher parental education levels.

### Q7: Grade Assignment
Assigning letter grades to scores using a custom function.

```python
def grade(x):
    if x>=90: return "A"
    elif x>=75: return "B"
    elif x>=60: return "C"
    elif x>=50: return "D"
    else: return "F"

grade_035_df = stu_035_df.copy()

for col in ["math score","reading score","writing score"]:
    grade_035_df[col+"_g"] = grade_035_df[col].apply(grade)

grade_035_df.set_index(["gender","race/ethnicity"]).head()
```

Letter grades were assigned based on score ranges with A for 90-100, B for 80-89, C for 70-79, D for 60-69, and F for below 60. Reading has the highest percentage of A grades while math has the highest percentage of F grades. Female students receive more A and B grades across all subjects compared to male students. This indicates that reading is the strongest subject overall while math is the most challenging.

### Q8: Index Alignment Analysis
Comparing student scores with class averages.

```python
avg_035 = stu_035_df[["math score","reading score","writing score"]].mean()

dif_035 = stu_035_df[["math score","reading score","writing score"]] - avg_035

dif_035.head()
```

When comparing each students scores against the class average, math has the highest number of students scoring significantly below average. Most of these underperforming students are male and did not complete the test preparation course. Reading and writing show fewer students below average. This suggests that math requires more targeted intervention and test preparation helps reduce underperformance.

### Q9: Preparation vs Non-Preparation
Comparing performance between prepared and unprepared students.

```python
prep_yes = stu_035_df[stu_035_df["test preparation course"]=="completed"]
prep_no = stu_035_df[stu_035_df["test preparation course"]=="none"]

gap_035 = prep_yes[["math score","reading score","writing score"]].mean() - \
            prep_no[["math score","reading score","writing score"]].mean()

gap_035
```

Students who completed test preparation scored 5.89 points higher in math, 7.58 points higher in reading, and 7.56 points higher in writing compared to those who did not. Reading and writing show the largest percentage improvement of about 11.5 percent while math shows about 9 percent improvement. This demonstrates that test preparation is effective across all subjects with the greatest impact on language-based subjects.

### Q10: Comprehensive Report
Building a multi-level performance summary using grouping techniques.

```python
rep_035_df = stu_035_df.groupby(
    ["gender","parental level of education","test preparation course"]
)[["math score","reading score","writing score"]].mean()

rep_035_df.unstack()
```

The comprehensive analysis of student performance reveals several key patterns. Reading is the strongest subject with the highest average scores and most A grades, while math shows the highest failure rate and most variation. Test preparation course completion improves scores by 6 to 8 points across all subjects, with the greatest impact on reading and writing. Female students outperform male students across all subjects, particularly in reading and writing. Parental education level strongly correlates with student performance, as students with college-educated parents achieve higher scores. The composite performance index shows that top performers excel consistently across all subjects and most completed test preparation. Overall, test preparation, gender, and parental education are the most significant factors influencing academic performance.
